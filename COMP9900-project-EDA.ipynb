## 1. Data Set Type and Missing Values
- **Data Set Type**: It is mainly of the numeric type (int & float), and the compound label is 'object'.
- **Missing Values**: The dataset contains 90 sampling sites and 5720 compounds and microorganisms, with no missing values.

## 2. Top 20 Sampling Points for Compounds and Microorganisms
- **Histogram Representation**: Find and plot the top 20 sampling points for both compounds and microorganisms.

## 3. Top 10 Compounds and Microorganisms
- **Identify Top 10**: List and analyze the top 10 compounds and microorganisms.

## 4. Analysis of the 20 Sampling Points with the Largest Compound Distribution
- **Distribution Analysis**: Analyze the 20 sampling points that have the largest distribution of the largest number of compounds.

## 5. Statistical Analysis
- **Statistical Measures**: Perform statistical analysis on the data set (e.g., mean, median, standard deviation).

- **Heat Map Representation**: Determine the correlation between sample points and present it using a heat map.

## 7. Principal Component Analysis (PCA)
- **PCA**: Conduct a principal component analysis,  including principal component loading information.
  - **Impact of PCA1**: Highlight that PCA1 has the greatest impact.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# The dataset is read in and converted to CSV format for processing.
file_path = './2022data and 2023data.txt'
data = pd.read_csv(file_path, delimiter="\t")

# Make sure all output is not omitted.
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.width', None)

data_cleaned = data.drop(0)
data_cleaned.reset_index(drop=True, inplace=True)
data_cleaned.iloc[:, 1:] = data_cleaned.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')

# View data types and missing values
print(data_cleaned.info())
print(data_cleaned.isnull().sum())

# Calculate the total number of bacteria at each site
location_totals = data_cleaned.iloc[:, 1:].sum()

# Visualize the total number of bacteria per site
plt.figure(figsize=(20, 10))
ax = location_totals.sort_values().tail(20).plot(kind='bar', color='skyblue')
plt.title('20 sampling points with the most compounds and microorganisms', fontsize=16)
plt.xlabel('Sampling Location', fontsize=14)
plt.ylabel('Total Compounds and microbial Counts', fontsize=14)
plt.xticks(rotation=90, fontsize=10)
plt.yticks(fontsize=12)
plt.grid(True)

for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=10)
plt.tight_layout()
plt.show()

# Calculate and find the 10 most abundant bacteria
bacteria_totals = data_cleaned.iloc[:, 1:].sum(axis=1)
bacteria_totals = pd.DataFrame({
    'Bacteria': data_cleaned.iloc[:, 0],
    'Total Count': bacteria_totals
})
top_bacteria = bacteria_totals.sort_values(by='Total Count', ascending=False).head(10)
print(top_bacteria)

# Plot the distribution of specific bacteria at various locations
bacteria_name = top_bacteria.iloc[0, 0]
bacteria_distribution = data_cleaned[data_cleaned['Unnamed: 0'] == bacteria_name].iloc[0, 1:]
plt.figure(figsize=(20, 10))
ax = bacteria_distribution.sort_values().tail(20).plot(kind='bar', color='green')
plt.title(f'Distribution of {bacteria_name} Across Sampling Locations', fontsize=16)
plt.xlabel('Sampling Location', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=90, fontsize=10)
plt.yticks(fontsize=12)
plt.grid(True)

for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=10)
plt.tight_layout()
plt.show()

# statistical description
print(data_cleaned.describe())

# data correlation
plt.figure(figsize=(30, 25))
sns.heatmap(data_cleaned.iloc[:, 1:].corr(), annot=True, fmt=".1f", cmap='coolwarm',
            annot_kws={"size": 10}, cbar_kws={"shrink": 0.75})
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=12)
plt.title('Correlation Heatmap', fontsize=20)
plt.show()

# standardized data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_cleaned.iloc[:, 1:].dropna(axis=1))

# Principal Component Analysis (PCA)
pca = PCA(n_components=2)
pca_result = pca.fit_transform(data_scaled)
pca_df = pd.DataFrame(pca_result, columns=['PCA1', 'PCA2'])

# principal component loading
loadings = pd.DataFrame(pca.components_.T, columns=['PCA1', 'PCA2'], index=data_cleaned.columns[1:])
print(loadings)

# PCA Results Visualization
plt.figure(figsize=(20, 15))
plt.scatter(pca_df['PCA1'], pca_df['PCA2'], alpha=0.8)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Result')
plt.grid(True)
plt.show()
```
